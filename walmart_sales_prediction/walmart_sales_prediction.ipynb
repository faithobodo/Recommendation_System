```python
# to identify input and target

input_cols, target_col = data.columns[:-1], data.columns[-1]
inputs_df, targets = data[input_cols].copy(), data[target_col].copy()
```

```python
input_cols
```

```python
# to sepearate dataset into numerical and categorical, and then casting into a list for normalizing and one hot encoding
numeric_cols = data[input_cols].select_dtypes(include=np.number).columns.tolist()
categorical_cols = data[input_cols].select_dtypes(include='object').columns.tolist()
```

```python
categorical_cols
```

```python
# to scale the numerical data
imputer = SimpleImputer().fit(inputs_df[numeric_cols])
inputs_df[numeric_cols] = imputer.transform(inputs_df[numeric_cols])
scaler = MinMaxScaler().fit(inputs_df[numeric_cols])
inputs_df[numeric_cols] = scaler.transform(inputs_df[numeric_cols])
```

```python
# for numerical variable, use one hot encoding to scale categorical variables, since the model can only be trained with numbers
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore').fit(inputs_df[categorical_cols])
encoded_cols = list(encoder.get_feature_names(categorical_cols))
inputs_df[encoded_cols] = encoder.transform(inputs_df[categorical_cols])
```

### Hyperparameter Tuning for randomforest regressor

```python
def test_params(**params):
    model = RandomForestRegressor(random_state=0, n_jobs=-1, n_estimators=16, **params).fit(train_inputs, train_targets)
    train_rmse = mean_squared_error(model.predict(train_inputs), train_targets, squared=False)
    test_rmse = mean_squared_error(model.predict(test_inputs), test_targets, squared=False)
    return train_rmse, test_rmse
```

```python
def test_param_and_plot(param_name, param_values):
    train_errors, test_errors = [], [] 
    for value in param_values:
        params = {param_name: value}
        train_rmse, test_rmse = test_params(**params)
        train_errors.append(train_rmse)
        test_errors.append(test_rmse)
    plt.figure(figsize=(10,6))
    plt.title('Overfitting curve: ' + param_name)
    plt.plot(param_values, train_errors, 'b-o')
    plt.plot(param_values, test_errors, 'r-o')
    plt.xlabel(param_name)
    plt.ylabel('RMSE')
    plt.legend(['Training', 'Validation'])
    plt.show()
```

```python
test_params()
```

```python
test_param_and_plot('min_samples_leaf', [1, 2, 3, 4, 5])
```

```python
test_params(min_samples_leaf = 5)
```

```python
test_param_and_plot('max_leaf_nodes', [20, 25, 30, 35, 40])
```

```python
test_params(max_leaf_nodes = 20)
```

```python
test_param_and_plot('max_depth', [5, 10, 15, 20, 25])
```

```python
test_params(max_depth = 10)
```
